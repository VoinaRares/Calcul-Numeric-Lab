```python
#EX2A:

import numpy as np

A = np.array([
    [10, 7, 8, 7],
    [7, 5, 6, 5],
    [8, 6, 10, 9],
    [7, 5, 9, 10]
], dtype=float)

b = np.array([32, 23, 33, 31], dtype=float)

A, b
```

```python
# Check symmetry
print("Symmetric:", np.allclose(A, A.T))

# Check positive definiteness via Cholesky
try:
    np.linalg.cholesky(A)
    print("Positive definite: True")
except np.linalg.LinAlgError:
    print("Positive definite: False")
```

```python
#B:

def cholesky(A):
    n = A.shape[0]
    L = np.zeros_like(A)

    for i in range(n):
        for j in range(i + 1):
            s = 0
            for k in range(j):
                s += L[i][k] * L[j][k]

            if i == j:
                L[i][j] = np.sqrt(A[i][i] - s)
            else:
                L[i][j] = (A[i][j] - s) / L[j][j]

    return L

L = cholesky(A)
print("L =\n", L)
```

```python
#C:

y = np.linalg.solve(L, b)

x = np.linalg.solve(L.T, y)

print("Solution x =", x)
```

```python
# Check Ax ≈ b
print("Ax =", A @ x)
print("b  =", b)
```
